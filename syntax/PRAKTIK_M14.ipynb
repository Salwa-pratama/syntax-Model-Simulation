Langkah 1: Persiapan Model dan Baseline

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import  LinearRegression
import pickle

# 1. Menyimpan data historis sederhana
#  Fitur: [Iklan (Juta), Diskon (%)]
n_samples = 500
# Variabel  kolm
iklan_random = np.random.randint(5, 51, n_samples) # Menambahkan nilai random anggaran iklan dari 5 jt - 50jt
diskon_random = np.random.randint(0, 51, n_samples) # diskon randm 0% - 50%
cabang_random = np.random.randint(0, 10, n_samples) # cabang 0 - 10 cabang

x_train = np.column_stack((iklan_random, diskon_random, cabang_random))

# 2. GENERATE TARGET (Y) MENGGUNAKAN LOGIKA BISNIS + NOISE
# Asumsi Logika:
# - Tiap 1 Cabang nambah untung dasar 30 Juta
# - Tiap 1 Juta Iklan nambah untung 2 Juta
# - Tiap 1% Diskon ngurangin untung 1.5 Juta
keuntungan_dasar = (cabang_random * 30) + (iklan_random * 2) - (diskon_random * 1.5)

# Bikin noise (acak) agar data tidak berbentuk garis lurus sempurna (biar realistis)
# np.random.normal(mean, standar_deviasi, jumlah_data)
noise = np.random.normal(0, 10, n_samples)

# Keuntungan akhir (y_train)
y_train = keuntungan_dasar + noise

# 2. Melatih model (mesin Replika)
model = LinearRegression().fit(x_train, y_train)

# Simpan model
with open ("../models/model2.pkl", "wb") as file:
    pickle.dump(model, file)

# 3. Menetapkan Skenario Dasar (Baseline)

# Kondisi saat ini : Iklan 10 juta , Diskon 10%, jumlah cabang 2
baseline_input = np.array([[10, 10, 2]])
baseline_pred = model.predict(baseline_input)[0]
print(f"Prediksi Keuntungan Baseline: Rp {baseline_pred:.2f} Juta")
```

Langkah 2 : Membangun logika simulator (Analisis What-if)

```python
def run_simulation(new_iklan, new_diskon):
    # Input baru dari user (Intervensi)
    intervention_input = np.array([[new_iklan, new_diskon]])

    # Prediksi hasil  intervensi
    prediction = model.predict(intervention_input)[0]

    # Menghitung Delta (Selisih)
    delta_y = prediction - baseline_pred

    return prediction, delta_y
```
